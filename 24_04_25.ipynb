```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

link='https://github.com/mike1502/HerramientasComputacionales/blob/main/Pandas/ebola_2014_2016_clean.csv'
code='?raw=true' #para que pandas lea el archivo csv
url=link+code #se unen el link y el codigo para leer el archivo
df=pd.read_csv(url,sep=',',index_col=1)

df.head() #visualizar el encabezado del dataframe (base de datos), por defecto muestra las primeras 5 filas y el encabezado
```

```python
df.tail() #visualiza las ultimas filas del dataframe
```

```python
df.info() #informacion y tipo de dato por columna
```

```python
df.describe() #visualizar las estadisticas de cada columna
```

```python
df['Country'].isin(['Mexico']).sum()
#consulto en el df, en la columna de pais, si existe informacion sobre el ebola en mexico
```

```python
df.isnull().sum()
#entro a la base de datos para buscar datos nulos y sumarlos
```

```python
promedio_muertos=df['No. of confirmed deaths'].mean()
df['No. of confirmed deaths'].fillna(promedio_muertos, inplace=True) #promediar los casos de muertes confirmadas y reemplazar los valores vacios con ese. alor.
```

```python
df.tail(10)
```

```python
df.isnull().sum()
```

```python
#vamos a contar los casos por pais
conteo=df.groupby('Country').size().reset_index(name='conteo')
conteo=conteo.sort_values(by='conteo',ascending=False)
conteo
```

```python
data_missing=df['No. of confirmed deaths'].isnull().groupby(df['Country']).sum()
data_missing=data_missing.sort_values(ascending=False)
data_missing
```

```python
plt.pie(conteo['conteo'],labels=conteo['Country'],autopct='%1.1f%%')
plt.axis('equal')
plt.title('Casos de Ebola por Pa√≠s')
plt.savefig('Casos de Ebola por Pa√≠s',dpi=300,format='png')
plt.show()
```

```python
Senegal=df[df['Country']=='Senegal']
Senegal.head()
```

```python
Senegal.isnull().sum() #comprobamos que no hay valores nulos
```

```python
Senegal.shape #cuantas filas y cuantas columnas hay
```

```python
Senegal_infectados=Senegal['No. of confirmed cases'][0:-1:10] #asignamos variable de infectados
Senegal_muertos=Senegal['No. of confirmed deaths'][0:-1:10] #asignamos variable para las muertes
fechas=Senegal.index[0:-1:10] #extraemos las fechas para el rango propuesto
plt.figure(figsize=(10,5))
plt.plot(fechas,Senegal_infectados,'o:r', ms=3, label='infectados')
plt.plot(fechas,Senegal_muertos, ms=3, label='muertos')
plt.grid() #malla de la grafica
plt.legend()
plt.xticks(rotation=90,fontsize=8)
plt.savefig('Casos de Ebola en Senegal',dpi=300,format='png')
plt.show
```

```python
Liberia=df[df['Country']=='Liberia']
Liberia.head()
```

```python
Liberia.isnull().sum()
```

```python
Liberia_hombres=Liberia['Male Cases'][0:-1:10]
Liberia_mujeres=Liberia['Female Cases'][0:-1:10]
fechas=Liberia.index[0:-1:10]
plt.figure(figsize=(10,5))
plt.plot(fechas,Liberia_hombres,'o:r', ms=3, label='hombres')
plt.plot(fechas,Liberia_mujeres, ms=3, label='mujeres')
plt.grid()
plt.legend()
plt.xticks(rotation=90,fontsize=8)
plt.savefig('Generos en casos de Ebola en Liberia',dpi=300,format='png')
plt.show
```

```python
MC=Liberia['Male Cases'][0:-1:10]
B=Liberia['Boys'][0:-1:10]
AM=Liberia['Adult Male'][0:-1:10]
OM=Liberia['Old Male'][0:-1:10]

fechas=Liberia.index[0:-1:10]

plt.figure(figsize=(10,5))
plt.bar(fechas,MC,color='blue',label='Total', alpha=0.6)
plt.bar(fechas,B,color='yellow', label='Ni√±os', alpha=0.6)
plt.bar(fechas,AM,color='pink' label='Adultos', alpha=0.6)
plt.bar(fechas,OM,color='red',label='Adulto Mayor', alpha=0.6)
```
